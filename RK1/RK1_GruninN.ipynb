Задача №24.
Для набора данных для одного (произвольного) числового признака проведите обнаружение и удаление выбросов на основе 5% и 95% квантилей.

```python
low_b = data['Weight'].quantile(0.05)
up_b = data['Weight'].quantile(0.95)

outliers_temp = np.where(data['Weight'] > up_b, True, 
                                 np.where(data['Weight'] < low_b, True, False))
data_trimmed = data.loc[~(outliers_temp), ]  
data_trimmed.shape
```
